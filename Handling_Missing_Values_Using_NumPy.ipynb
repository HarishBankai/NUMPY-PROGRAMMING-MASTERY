```python
import numpy as np

# Create a sample 1D array with some NaN values
data1 = np.array([86, 77, 75, 78, 82, 95, np.nan, 80, np.nan, 68, np.nan, 95, 96, 50, np.nan, 85, np.nan, 52, 77])

# Calculate the mean of the array, ignoring NaN values
mean = np.nanmean(data1)
print("MEAN OF 1D ARRAY : ", mean)

# Calculate the median of the array, ignoring NaN values
median = np.nanmedian(data1)
print("MEDIAN OF 1D ARRAY : ", median)

# Print the original 1D array and a boolean mask showing where NaN values are located
print("1D ARRAY NULL VALUE POSITIONS : \n", data1)
print(np.isnan(data1))

# Create a sample 2D array with some NaN values
data2 = np.array([[6, 7, np.nan, 9] ,
                  [np.nan, 4, 15, np.nan] ,
                  [np.nan, 10, np.nan, np.nan] ,
                  [5, np.nan, 89, np.nan]])

# Print the original 2D array and a boolean mask showing where NaN values are located
print("\n2D ARRAY NULL VALUES POSITIONS: \n", data2)
print(np.isnan(data2))

# Replace NaN values in the 1D array with the calculated mean
data1_replaced = np.nan_to_num(data1, nan=mean)
print("After replacing the NaN values in 1D array : ", data1_replaced)

# Replace NaN values in the 1D array with the calculated median
data1_replaced = np.nan_to_num(data1, nan=median)
print("After replacing the NaN values in 1D array : ", data1_replaced)

# Remove NaN values in the 1D array
data1_remove = data1[~np.isnan(data1)]
print("After removing the NaN values in 1D array : ", data1_remove)

# Remove NaN values in the 2D array
data2_remove = data2[~np.isnan(data2)]
print("After removing the NaN values in 2D array : ", data2_remove)
```
